```python
def read_weather_desc(desc):
    thunderstorm = []
    thunderstorm_prox = []
    drizzle = []
    rain = []
    freezing_rain = []
    rain_prox = []
    snow = []
    sleet = []
    rain_snow = []
    atmopsphere1 = []
    atmosphere2 = []
    squall = []
    tornado = []
    clouds = []
    for single_desc in desc:
        thunderstorm1, thunderstorm_prox1, drizzle1, rain1, freezing_rain1, rain_prox1, snow1, sleet1, rain_snow1, atmopsphere11, atmosphere21, squall1, tornado1, clouds1 = read_weather_desc_single_day(single_desc)
        thunderstorm.append(thunderstorm1)
        thunderstorm_prox.append(thunderstorm_prox1)
        drizzle.append(drizzle1)
        rain.append(rain1)
        freezing_rain.append(freezing_rain1)
        rain_prox.append(rain_prox1)
        snow.append(snow1)
        sleet.append(sleet1)
        rain_snow.append(rain_snow1)
        atmopsphere1.append(atmopsphere11)
        atmosphere2.append(atmosphere21)
        squall.append(squall1)
        tornado.append(tornado1)
        clouds.append(clouds1)
    return thunderstorm, thunderstorm_prox, drizzle, rain, freezing_rain, rain_prox, snow, sleet, rain_snow, atmopsphere1, atmosphere2, squall, tornado, clouds

def read_weather_desc_single_day(desc):

    thunderstorm = 0
    thunderstorm_prox = 0
    drizzle = 0
    rain = 0
    freezing_rain = 0
    rain_prox = 0
    snow = 0
    sleet = 0
    rain_snow = 0
    atmopsphere1 = 0
    atmosphere2 = 0
    squall = 0
    tornado = 0
    clouds = 0

    if isinstance(desc, (float)) and math.isnan(desc):
        return thunderstorm, thunderstorm_prox, drizzle, rain, freezing_rain, rain_prox, snow, sleet, rain_snow, atmopsphere1, atmosphere2, squall, tornado, clouds

    desc = desc.lower()

    # thunderstorm
    if desc == 'thunderstorm with light drizzle'.lower():
        thunderstorm = 1
    if desc == 'thunderstorm with drizzle'.lower():
        thunderstorm = 2
    if desc == 'thunderstorm with heavy drizzle'.lower():
        thunderstorm = 3
    if desc == 'thunderstorm with light rain'.lower():
        thunderstorm = 4
    if desc == 'thunderstorm with rain'.lower():
        thunderstorm = 5
    if desc == 'thunderstorm with heavy rain'.lower():
        thunderstorm = 6
    if desc == 'light thunderstorm'.lower():
        thunderstorm = 7
    if desc == 'thunderstorm'.lower():
        thunderstorm = 8
    # thunderstorm prox
    if desc == 'proximity thunderstorm'.lower():
        thunderstorm_prox = 1
    if desc == 'proximity thunderstorm with drizzle'.lower():
        thunderstorm_prox = 2
    if desc == 'proximity thunderstorm with rain'.lower():
        thunderstorm_prox = 3
    # drizzle
    if desc == 'light intensity drizzle'.lower():
        drizzle = 1
    if desc == 'drizzle'.lower():
        drizzle = 2
    if desc == 'heavy intensity drizzle'.lower():
        drizzle = 3
    if desc == 'light intensity drizzle rain'.lower():
        drizzle = 4
    if desc == 'shower drizzle'.lower():
        drizzle = 5
    #rain
    if desc == 'light rain'.lower():
        rain = 1
    if desc == 'moderate rain'.lower():
        rain = 2
    if desc == 'heavy intensity rain'.lower():
        rain = 3
    if desc == 'very heavy rain'.lower():
        rain = 4
    if desc == 'light intensity shower rain'.lower():
        rain = 5
    if desc == 'shower rain'.lower():
        rain = 6
    if desc == 'heavy intensity shower rain'.lower():
        rain = 7
    if desc == 'ragged shower rain'.lower():
        rain = 8
    #FreezingRain
    if desc == 'freezing rain'.lower():
        freezing_rain = 1
    #RainProx
    if desc == 'proximity moderate rain'.lower():
        rain_prox = 1
    if desc == 'proximity shower rain'.lower():
        rain_prox = 2
    #Snow
    if desc == 'light snow'.lower():
        snow = 1
    if desc == 'Snow'.lower():
        snow = 2
    if desc == 'Heavy snow'.lower():
        snow = 3
    #Sleet
    if desc == 'Sleet'.lower():
        sleet = 1
    if desc == 'Light shower sleet'.lower():
        sleet = 2
    #RainSnow
    if desc == 'Light rain and snow'.lower():
        rain_snow = 1
    if desc == 'Light shower snow'.lower():
        rain_snow = 2
    if desc == 'Shower snow'.lower():
        rain_snow = 3
    if desc == 'Heavy shower snow'.lower():
        rain_snow = 4
    #Atmopsphere1 ?
    if desc == 'mist'.lower():
        atmopsphere1 = 1
    if desc == 'Smoke'.lower():
        atmopsphere1 = 2
    if desc == 'Haze'.lower():
        atmopsphere1 = 3
    if desc == 'fog'.lower():
        atmopsphere1 = 4
    #Atmosphere2 ?
    if desc == 'sand'.lower():
        atmosphere2 = 1
    if desc == 'dust'.lower():
        atmosphere2 = 2
    if desc == 'volcanic ash'.lower():
        atmosphere2 = 3
    if desc == 'proximity sand/dust whirls'.lower():
        atmosphere2 = 4
    if desc == 'sand/dust whirls'.lower():
        atmosphere2 = 5
    #Squall ?
    if desc == 'squalls'.lower():
        squall = 1
    #Tornado ?
    if desc == 'tornado'.lower():
        tornado = 1
    #Clouds
    if desc == 'sky is clear'.lower():
        clouds = 1
    if desc == 'few clouds'.lower():
        clouds = 2
    if desc == 'scattered clouds'.lower():
        clouds = 3
    if desc == 'broken clouds'.lower():
        clouds = 4
    if desc == 'overcast clouds'.lower():
        clouds = 5

    return thunderstorm, thunderstorm_prox, drizzle, rain, freezing_rain, rain_prox, snow, sleet, rain_snow, atmopsphere1, atmosphere2, squall, tornado, clouds
```
